```python
 1# Tasas de cambio fijas
tasa_dolar_a_pesos = 18.00       # 1 USD = 18.00 MXN
tasa_dolar_a_euros = 0.85        # 1 USD = 0.85 EUR
tasa_dolar_a_libras = 0.75       # 1 USD = 0.75 GBP
tasa_dolar_a_yuan = 6.50         # 1 USD = 6.50 CNY
tasa_dolar_a_rublo = 80.00       # 1 USD = 80.00 RUB

# Solicitar la cantidad en dólares al usuario
dolares = 500
# Calcular las conversiones
pesos = dolares * tasa_dolar_a_pesos
euros = dolares * tasa_dolar_a_euros
libras = dolares * tasa_dolar_a_libras
yuan = dolares * tasa_dolar_a_yuan
rublo = dolares * tasa_dolar_a_rublo

# Mostrar las tasas de cambio
print("\Tasas de cambio:")
print(f"1 USD = {tasa_dolar_a_pesos} MXN")
print(f"1 USD = {tasa_dolar_a_euros} EUR")
print(f"1 USD = {tasa_dolar_a_libras} GBP")
print(f"1 USD = {tasa_dolar_a_yuan} CNY")
print(f"1 USD = {tasa_dolar_a_rublo} RUB")

# Imprimir los resultados de las conversiones
print("\conversion:")
print(f"{dolares} USD = {pesos:.2f} MXN") #el 2f es para mostrar 2 decimales que se utiliza para resultados financieros
print(f"{dolares} USD = {euros:.2f} EUR")
print(f"{dolares} USD = {libras:.2f} GBP")
print(f"{dolares} USD = {yuan:.2f} CNY")
print(f"{dolares} USD = {rublo:.2f} RUB")
```

```python
#6 lista de numeros
def separar_numeros(numeros):
    positivos = []
    negativos = []


    for numero in numeros:
        if numero > 0:
            positivos.append(numero)
        elif numero < 0:
            negativos.append(numero)

    return positivos, negativos


numeros =[5,6,8,-23,12,-7,-12]
positivos, negativos = separar_numeros(numeros)
print(f"positivos: {positivos}")
print(f"negativos: {negativos}")
```

```python
#7 lista de datos
def filtrar_datos(datos):
  enteros = []
  floats = []
  strings = []
  booleanos = []

  for dato in datos:
    if isinstance(dato, int):  #isinstance nos sirve para verificar si un objeto pertenece a un cierto grupo
      enteros.append(dato)
    elif isinstance(dato, float):
      floats.append(dato)
    elif isinstance(dato, str):
      strings.append(dato)
    elif isinstance(dato, bool):
      booleanos.append(dato)

  return enteros, floats, strings, booleanos

Datos= [5, 2.4, "hola", 7, 10, 3.14,True,False, "Hi", 0, 0.265]
enteros, floats, strings, booleanos = filtrar_datos(Datos)
print(f"enteros: {enteros}")
print(f"floats: {floats}")
print(f"strings: {strings}")
print(f"booleanos: {booleanos}")
```

```python
#9 factorial
def factorial():
    while True:
        entrada = input("ingresar numero natural")

        if entrada.isdigit():
            num = int(entrada)
            break

        else:
            print("entrada invalida. debe ser un numero natural")

    n = num
    factorial_result = 1

    for i in range(1, n + 1):
        factorial_result *= i

    print(f"el factorial de {n} es {factorial_result}")
    return factorial_result

factorial()
```
